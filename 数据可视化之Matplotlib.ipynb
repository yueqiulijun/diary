## 数据可视化之Matplotlib
Matplotlib是一个用于Python中绘制数组的2D图形库。它起源于模仿MATLAB的图形命令，但现如今它独立于MATLAB，可以以Pythonic和面向对象的方式使用。Matplotlib大量使用NumPy和其他扩展代码，即使对于大型数组也能提供良好的性能。

```python
import matplotlib.pyplot as plt  
plt.plot([1,2,3,4])  
plt.ylabel('some number')  
plt.show()  
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
fig,axes = plt.subplots(  
    nrows=2,       #坐标轴行数  
    ncols=2,       #坐标轴列数  
    sharex=False,  #共享x轴  
    sharey=False)  #共享y轴  
#选定0，0即第一行第一个坐标轴做直方图  
axes[0,0].hist(np.random.randint(0,20,100),bins=10,color='r',alpha=0.5)  
#选定0，1即第一行第二个坐标轴做饼图  
x1=np.random.randint(1,20,6)  
labels=list('ABCDEF')  
explode=(0,)*5+(0.1,)  
axes[0,1].pie(x1,explode=explode,labels=labels,autopct='%1.0f%%')  
#选定1，0即第二行第一个坐标轴做曲线图  
x2=np.arange(0,2*np.pi,0.02)  
y2=np.sin(np.arange(0,2*np.pi,0.02))  
axes[1,0].plot(x2,y2,'r--')  
#选定1，1即第二行第二个坐标轴做条形图  
x3=np.arange(0,5)  
y3=np.random.randint(10,100,5)  
axes[1,1].bar(x3,y3)  
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
N = 50  
x = np.random.rand(N)  
y = np.random.rand(N)  
colors = np.random.rand(N)  
area = np.pi*(15*np.random.rand(N))**2  
plt.scatter(x,          #x  
            y,          #y  
            s=area,     #点的大小  
            c=colors,   #颜色  
            alpha=0.5)  #透明度  
plt.title('scatter')    #设置标题  
plt.xlabel('x')         #设置x轴标签  
plt.ylabel('y')         #设置y轴标签  
plt.show()  
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
np.random.seed(0)  
N = 20  
theta = np.linspace(0.0,2 * np.pi,N,endpoint=False)   #x值  
radii = 10 * np.random.rand(N)                       #y值  
width = np.pi / 4 * np.random.rand(N)                   #条状图宽度  
ax = plt.subplot(111,projection='polar')            #定义极坐标轴  
bars = ax.bar(theta,radii,width=width,bottom=0.0)   #画条形图                       
#自定义颜色及透明度  
for r,bar,in zip(radii,bars):  
    bar.set_facecolor(plt.cm.viridis(r/10.))  
    bar.set_alpha(0.5)  
plt.show()   
```
